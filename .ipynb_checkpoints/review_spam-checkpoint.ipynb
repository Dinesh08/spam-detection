```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import string
import nltk
from nltk.stem import SnowballStemmer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,precision_recall_fscore_support,precision_score,recall_score,f1_score
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
review_train = pd.read_csv("amazon-review-scraper-labelled1.csv",encoding="ISO-8859-1")
review_df=pd.read_csv("review_with_product.csv")
```

```python
review_train=review_train.fillna('')
review_df=review_df.fillna('')
```

```python
review_train.label = review_train.label.str.replace(' ', '')
```

```python
print("\n\n no of null values in each column \n",review_train.isnull().sum())
```

```python
def pre_process(text):

    text = text.translate(str.maketrans('', '', string.punctuation))
    text = [word for word in text.split() if word.lower() not in stopwords.words('english')]
    words = ""
    for i in text:
            stemmer = SnowballStemmer("english")
            words += (stemmer.stem(i))+" "
    return words
```

```python
textFeatures = review_train['content'].copy()
textFeatures = textFeatures.apply(pre_process)
textFeatures
```

```python
vectorizer = TfidfVectorizer("english")
features = vectorizer.fit_transform(textFeatures)
```

```python
op_textFeatures = review_df['review'].copy()
op_textFeatures = op_textFeatures.apply(pre_process)
op_features = vectorizer.transform(op_textFeatures)
```

```python
mnb = MultinomialNB(alpha=0.2)
mnb.fit(features,review_train['label'])
prediction = mnb.predict(op_features)
print("\nMNB Prediction:\n",prediction)
```

```python
svc = SVC(kernel='sigmoid', gamma=1.0)
svc.fit(features,review_train['label'])
prediction = svc.predict(op_features)
print("\nSVC Prediction\n",prediction)
```

```python
features_train, features_test, labels_train, labels_test = train_test_split(features, review_train['label'], test_size=0.3, random_state=111)
```

```python
mnb1 = MultinomialNB(alpha=0.2)
mnb1.fit(features_train,labels_train)
```

```python
prediction1 = mnb1.predict(features_test)
print("accuracy ",accuracy_score(labels_test,prediction1))
print("precision_score spam ",precision_score(labels_test,prediction1,pos_label='spam'))
print("recall_score spam ",recall_score(labels_test,prediction1,pos_label='spam'))
print("f1_score spam ",f1_score(labels_test,prediction1,pos_label='spam'))
print("precision_score quality ",precision_score(labels_test,prediction1,pos_label='quality'))
print("recall_score quality",recall_score(labels_test,prediction1,pos_label='quality'))
print("f1_score quality ",f1_score(labels_test,prediction1,pos_label='quality'))
print(confusion_matrix(labels_test, prediction1))
print(classification_report(labels_test, prediction1))
```

```python
svc1 = SVC(kernel='sigmoid', gamma=1.0)
svc1.fit(features_train,labels_train)
```

```python
prediction2 = svc1.predict(features_test)
print("accuracy score spam ",accuracy_score(labels_test,prediction2))
print("precision_score spam ",precision_score(labels_test,prediction2,pos_label='spam'))
print("recall_score spam ",recall_score(labels_test,prediction2,pos_label='spam'))
print("f1_score spam ",f1_score(labels_test,prediction2,pos_label='spam'))
print("precision_score quality ",precision_score(labels_test,prediction2,pos_label='quality'))
print("recall_score quality ",recall_score(labels_test,prediction2,pos_label='quality'))
print("f1_score quality ",f1_score(labels_test,prediction2,pos_label='quality'))
print(confusion_matrix(labels_test, prediction2))
print(classification_report(labels_test, prediction2))
```

```python
objects = ("multinomialNB","svc")
y_pos = np.arange(len(objects))
performance = [accuracy_score(labels_test,prediction1),accuracy_score(labels_test,prediction2)]
#performance_test=[NBTest,RFTest,TreeTest,pred_test]
plt.figure(figsize=(3,5))
plt.bar(objects, performance,width=0.3,align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Accuracy')
plt.title('Performance of Models')
plt.show()
```
